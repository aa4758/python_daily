라인 단위로 읽기

```python
file = open('파일열기.ipynb','r', encoding = 'utf8')
while True:
    line = file.readline()
    if not line:
        break
#    line = line.strip()  줄바꿈 하지 않고 한 줄로 출력
    print(line, end="")
file.close()
```

```python
with open('파일열기.ipynb','r', encoding = 'utf8') as file:  #file.close()할 필요 없음
    lines = file.readlines()
    for line in lines:
        print(f'{line}', end="")
    
```

```python
with open('sample.txt','a', encoding='utf8') as file:
    file.write('어.. 재밌다\n')
    file.write('헤헤헤\n')
```

```python
with open('odd_number.txt','w', encoding='utf8') as file:
    for i in range(1, 30):
        if i % 2 == 1:
            file.write('%d\n' %i)
```

```python
with open('even_number.txt','w', encoding='utf8') as file:
    for i in range(1, 30):
        if i % 2 == 0:
            file.write('%d\n' %i)
```

```python
result = []

with open('파일열기.ipynb','r', encoding='utf8') as file:
    for line in file.readlines():
        word_list = list(line)
        for w in word_list:
            result.append(w)
print('파일 단어수 {}개'.format(len(result)))
```

```python
result = []

with open('파일열기.ipynb','r', encoding='utf8') as file:
    for line in file.readlines():
        word_list = list(line)
        for _ in range(len(word_list)):
            result.append(word_list.pop())
            
print('파일 단어수 {}개'.format(len(result)))
```
