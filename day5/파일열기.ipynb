````python
result = []

with open('파일열기.ipynb','r', encoding='utf8') as file:
    for line in file.readlines():
        word_list = list(line)
        for w in word_list:
            result.append(w)
print('파일 단어수 {}개'.format(len(result)))
````

````python
result = []

with open('파일열기.ipynb','r', encoding='utf8') as file:
    for line in file.readlines():
        word_list = list(line)
        for _ in range(len(word_list)):
            result.append(word_list.pop())
            
print('파일 단어수 {}개'.format(len(result)))
````

### `try ... finally`

에러와 관계없이 항상 마지막에 실행되어야 할 경우:

````python
file = open('tttt.txt', 'r')

try:
    file.readlines()
finally:
    file.close()
````

````python
try:
    file = open('tttt.txt', 'r')
    file.readlines()
except FileNotFoundError:
    pass
````
