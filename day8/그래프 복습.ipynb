```python
import csv
import matplotlib.pyplot as plt
from matplotlib import rc

f = open('cafe_2year.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

xdata = []
ydata = []
count = 1

for line in lines:
    if count%3 == 0 :
        xdata.append(line[0])
        ydata.append(line[1])

    count += 1
```

```python
plt.bar(xdata, ydata)

font1 = {'size' :18, 'color' : 'green'}

plt.title('전국의 월별 카페수', fontdict = font1)
plt.xlabel('년월')
plt.ylabel('카페수')
plt.show()
```

```python
import csv

f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

xdata = []
ydata1 = []
ydata2 = []
ydata3 = []
ydata4 = []

for line in lines:
    xdata.append(line[0])
    ydata1.append(int(line[1]))
    ydata2.append(int(line[2]))
    ydata3.append(int(line[3]))
    ydata4.append(int(line[4]))
    
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='Malgun Gothic')

plt.plot(xdata, ydata1, label = '일반의', linestyle = '--', marker = 's')
plt.plot(xdata, ydata2, label = '인턴', linestyle = '-', marker = '*')
plt.plot(xdata, ydata3, label = '레지던트', linestyle = ':', marker = 'o')
plt.plot(xdata, ydata4, label = '전문의', marker = 'd')
plt.legend(loc='upper center')

plt.title('서울 및 광역시 종합병원 의사 수', fontdict = font1)
plt.legend(loc = 'best')

plt.show()
```

```python
import csv

f = open('doctor2.csv', 'r', encoding='utf-8')
lines = csv.reader(f)

header = next(lines)

kind = ['일반의', '인턴', '레지던트', '전문의']

numbers = []
seoul = next(lines)

total = 0
for i in range(4) :
    total += int(seoul[i+1])
    
for i in range(4) :
    persent = int(seoul[i+1])/total;
    numbers.append(persent)
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

plt.pie(numbers, explode=(0, 0.1, 0, 0), labels = kind, autopct= '%.1f%%', shadow=True)
plt.title('서울 종합병원 의사 유형별 분포', fontdict = font1)

plt.show()
```
